```python
# Get Dataset
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
from scipy import stats
import plotly
```

```python
os.getcwd()
```

```python
raw_data = pd.read_csv('./Crossover_TakeHome_Data.csv')
```

```python
import re
raw_data['updated_ModalityDSC'] = raw_data["ModalityDSC"].map(lambda x: re.sub(r'\W+', '', x))
```

```python
# drop NAs
data = raw_data.dropna(subset = ["GenderCD","RiskNBR"])
```

```python
import plotly.express as px



# Create the Plotly Express bar plot
fig = px.bar(data, x="updated_ModalityDSC", y="VisitsCNT", color="ClinicID",
             title="Count of Patient Visits by Modality for Each Clinic Location",
             labels={"updated_ModalityDSC": "Modality Type", "VisitsCNT": "Patient Visits"},
             category_orders={"updated_ModalityDSC": ["A", "B", "C"]},  # Specify order of x-axis categories
             width=800, height=400)

# Show the plot
fig.show()
```
